#1. Análise do faturamento


```python
import pandas as pd

# Caminho dos arquivos
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

# Dados para carregar
loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)


print('Dados Carregados')
```

```python
# Criar coluna "Receita" para cada arquivo/loja
for loja in [loja1, loja2, loja3, loja4]:
    loja['Receita'] = loja['Preço'] + loja['Frete']

# Faturamento total de cada loja
faturamento_lojas = {
    "Loja 1": loja1['Receita'].sum(),
    "Loja 2": loja2['Receita'].sum(),
    "Loja 3": loja3['Receita'].sum(),
    "Loja 4": loja4['Receita'].sum()
}

# Faturamento com formatação
print("Faturamento total por loja:\n")
for loja, valor in faturamento_lojas.items():
    valor_formatado = f"R$ {valor:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")
    print(f"{loja}: {valor_formatado}")
```

# 2. Vendas por Categoria


```python
import pandas as pd

# Carregando os dados
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
]

df = pd.concat([pd.read_csv(url) for url in urls], ignore_index=True)

# Quantidade de vendas por categoria

categorias_mais_vendidas = df['Categoria do Produto'].value_counts()
categoria_top_vendas = categorias_mais_vendidas.idxmax()
vendas_top = categorias_mais_vendidas.max()

print("Categorias mais vendidas (por quantidade):\n")
print(categorias_mais_vendidas)
print(f"\n A categoria mais vendida é: _{categoria_top_vendas}_ com {vendas_top} vendas.")
```

```python
df['Receita'] = df['Preço'] + df['Frete']
faturamento_por_categoria = df.groupby('Categoria do Produto')['Receita'].sum().sort_values(ascending=False)
categoria_top_faturamento = faturamento_por_categoria.idxmax()
faturamento_top = faturamento_por_categoria.max()

# Formatar R$ bonitinho
faturamento_top_str = f"R$ {faturamento_top:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")

print("\n Faturamento total por categoria:\n")
print(faturamento_por_categoria.apply(lambda x: f"R$ {x:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")))
print(f"\n A categoria com maior faturamento é: _{categoria_top_faturamento}_ com {faturamento_top_str}.")
```

# 3. Média de Avaliação das Lojas

```python
import pandas as pd

# Carregando os dados
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
]

# Marcar de qual loja
lojas = []
for i, url in enumerate(urls, start=1):
    df_loja = pd.read_csv(url)
    df_loja['Loja'] = f'Loja {i}'
    lojas.append(df_loja)

df = pd.concat(lojas, ignore_index=True)


# Média geral das avaliações

media_geral = df['Avaliação da compra'].mean()
print(f" Média geral de avaliação dos clientes: {media_geral:.2f}")

#  Média por loja

avaliacoes_por_loja = df.groupby('Loja')['Avaliação da compra'].mean().sort_values(ascending=False)
print("\n Avaliação média por loja:\n")
print(avaliacoes_por_loja.round(2))

# Melhor loja
melhor_loja = avaliacoes_por_loja.idxmax()
nota_melhor = avaliacoes_por_loja.max()

# Pior loja
pior_loja = avaliacoes_por_loja.idxmin()
nota_pior = avaliacoes_por_loja.min()

print(f"\n A loja com melhor avaliação média é:_{melhor_loja}_ com nota {nota_melhor:.2f}.")
print(f" A loja com pior avaliação média é: _{pior_loja}_ com nota {nota_pior:.2f}.")
```

# 4. Produtos Mais e Menos Vendidos

```python
# Contar quantidade de vendas por produto
produtos_vendidos = df['Produto'].value_counts()

# Produto mais vendido
produto_mais_vendido = produtos_vendidos.idxmax()
qtd_mais_vendido = produtos_vendidos.max()

# Produto menos vendido
produto_menos_vendido = produtos_vendidos.idxmin()
qtd_menos_vendido = produtos_vendidos.min()

print(" Produtos mais vendidos (quantidade de vendas):\n")
print(produtos_vendidos.head(10))

print(f"\n O produto mais vendido é: _{produto_mais_vendido}_ com {qtd_mais_vendido} vendas.")
print(f" O produto menos vendido é: _{produto_menos_vendido}_ com {qtd_menos_vendido} venda(s).")
```

```python
# Garantir que a coluna de Receita existe
df['Receita'] = df['Preço'] + df['Frete']

# Faturamento por produto
faturamento_produtos = df.groupby('Produto')['Receita'].sum().sort_values(ascending=False)

# Produto que mais faturou
produto_mais_faturou = faturamento_produtos.idxmax()
valor_mais_faturou = faturamento_produtos.max()

# Produto que menos faturou
produto_menos_faturou = faturamento_produtos.idxmin()
valor_menos_faturou = faturamento_produtos.min()

# Formatar valores
formatar = lambda x: f"R$ {x:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")

print("\n Produtos com maior faturamento:\n")
print(faturamento_produtos.head(10).apply(formatar))

print(f"\n O produto com maior faturamento é: _{produto_mais_faturou}_ com {formatar(valor_mais_faturou)}.")
print(f" O produto com menor faturamento é: _{produto_menos_faturou}_ com {formatar(valor_menos_faturou)}.")
```

# 5. Frete Médio por Loja


```python
import pandas as pd

# Carregando os dados
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
]

# Marcar de qual loja
lojas = []
for i, url in enumerate(urls, start=1):
    df_loja = pd.read_csv(url)
    df_loja['Loja'] = f'Loja {i}'
    lojas.append(df_loja)

df = pd.concat(lojas, ignore_index=True)


# Frete médio por loja

frete_medio_por_loja = df.groupby('Loja')['Frete'].mean().sort_values()

print(" Frete médio por loja:\n")
print(frete_medio_por_loja.round(2))

# Loja com melhor (menor) frete
melhor_frete_loja = frete_medio_por_loja.idxmin()
melhor_frete_valor = frete_medio_por_loja.min()

# Loja com pior (maior) frete
pior_frete_loja = frete_medio_por_loja.idxmax()
pior_frete_valor = frete_medio_por_loja.max()

# Formatar valor
formatar = lambda x: f"R$ {x:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")

print(f"\n A loja com _melhor (menor) frete médio é: _{melhor_frete_loja}_ com frete médio de {formatar(melhor_frete_valor)}.")
print(f" A loja com pior (maior) frete médio é:_{pior_frete_loja}_ com frete médio de {formatar(pior_frete_valor)}.")
```

# Gráficos

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# URLs das lojas
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
]

# Função de formatação monetária
formatar_reais = lambda x: f'R$ {x:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.')

# Processar os dados de cada loja
dados_lojas = []

for i, url in enumerate(urls, start=1):
    df = pd.read_csv(url)
    df['Receita'] = df['Preço'] + df['Frete']
    loja = f'Loja {i}'

    dados_lojas.append({
        'Loja': loja,
        'Faturamento': df['Receita'].sum(),
        'Avaliação Média': df['Avaliação da compra'].mean(),
        'Frete Médio': df['Frete'].mean(),
        'Quantidade de Vendas': len(df)
    })

df_lojas = pd.DataFrame(dados_lojas)

# Gráficos
sns.set(style="whitegrid")
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('📊 Comparativo de Eficiência entre Lojas', fontsize=16, fontweight='bold')

def adicionar_rótulos(ax, valores, prefixo=""):
    for i, v in enumerate(valores):
        ax.text(i, v + v * 0.01, f"{prefixo}{v:,.2f}" if isinstance(v, float) else str(v),
                ha='center', va='bottom', fontsize=10, fontweight='bold')

# Faturamento
sns.barplot(data=df_lojas, x='Loja', y='Faturamento', ax=axs[0, 0], palette='Blues_d')
adicionar_rótulos(axs[0, 0], df_lojas['Faturamento'], prefixo="R$ ")
axs[0, 0].set_title(' Faturamento Total')
axs[0, 0].set_ylabel('Valor (R$)')

# Avaliação
sns.barplot(data=df_lojas, x='Loja', y='Avaliação Média', ax=axs[0, 1], palette='Greens_d')
adicionar_rótulos(axs[0, 1], df_lojas['Avaliação Média'])
axs[0, 1].set_title(' Avaliação Média dos Clientes')
axs[0, 1].set_ylim(0, 5)

# Frete médio
sns.barplot(data=df_lojas, x='Loja', y='Frete Médio', ax=axs[1, 0], palette='Oranges_d')
adicionar_rótulos(axs[1, 0], df_lojas['Frete Médio'], prefixo="R$ ")
axs[1, 0].set_title(' Frete Médio')
axs[1, 0].set_ylabel('Valor (R$)')

# Quantidade de vendas
sns.barplot(data=df_lojas, x='Loja', y='Quantidade de Vendas', ax=axs[1, 1], palette='Purples_d')
adicionar_rótulos(axs[1, 1], df_lojas['Quantidade de Vendas'])
axs[1, 1].set_title(' Quantidade de Vendas')

# Layout
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
